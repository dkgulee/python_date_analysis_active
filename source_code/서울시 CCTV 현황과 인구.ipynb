# pandas 기초 익히기
#### pandas의 데이터 유형 중 기초가 되는 것이 Series입니다. 대갈호로 만드는 파이썬 list 데이터로 만들 수 있습니다.

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([1,3,5,np.nan,6,8])
s
```

### pandas에는 날자형 데이터인 date_range가 있다. 기본 날짜를 지정하고 periods 옵션으로 7일간이라고 지정

```python
dates = pd.date_range('20130101', periods=7)
dates
```

## DataFrame 유형의 데이터를 만듬 random 변수와 인덱스를 dates로 지정 컬럼명은  A,B,C,D

```python
df = pd.DataFrame(np.random.randn(7,4), index=dates, columns=['A','B','C','D'])
df
```

```python
print(df.head(3))
print(df.tail(3))
print(df.info())
print(df.index)
print(df.columns)
print(df.values)
print(df.describe())
```
